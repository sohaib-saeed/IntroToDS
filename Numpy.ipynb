### Questin 01
Scientific Data Analysis with NumPy You are a scientist conducting research on climate change. You have collected temperature data from weather stations around the world stored in a NumPy array. Your task is to analyze the data using NumPy to:
1. Calculate the average temperature for each month across all stations.
2. Identify the station with the highest and lowest average temperature.
3. Determine the month with the largest temperature variation.
4. Visualize the temperature data using line plots or heatmaps.



<b>Use the following data set

```python
import numpy as np

# Sample temperature data for each month (in Celsius)
temperature_data = np.array([
    # January
    [10, 12, 15, 8, 9],
    # February
    [11, 13, 14, 9, 10],
    # March
    [12, 14, 16, 10, 11],
    # April
    [14, 16, 18, 12, 13],
    # May
    [16, 18, 20, 14, 15],
    # June
    [18, 20, 22, 16, 17],
    # July
    [20, 22, 24, 18, 19],
    # August
    [22, 24, 26, 20, 21],
    # September
    [20, 22, 24, 18, 19],
    # October
    [18, 20, 22, 16, 17],
    # November
    [15, 17, 19, 13, 14],
    # December
    [12, 14, 16, 10, 11]
])

# Station names for reference
station_names = ['Station A', 'Station B', 'Station C', 'Station D', 'Station E']
```

```python
#Average temperature for each month

monthly_avg_temp = np.mean(temperature_data, axis=1)
print("Average temperature for each month:", monthly_avg_temp)
```

```python
# Calculate average temperature for each station
station_avg_temp = np.mean(temperature_data, axis=0)
station_names = ['Station A', 'Station B', 'Station C', 'Station D', 'Station E']

# Identify the station with the highest and lowest average temperature
max_temp_station = station_names[np.argmax(station_avg_temp)]
min_temp_station = station_names[np.argmin(station_avg_temp)]

print("Station with the highest average temperature:", max_temp_station)
print("Station with the lowest average temperature:", min_temp_station)
```

```python
#Month with the largest temperature variation

# Calculate the standard deviation for each month
monthly_temp_variation = np.std(temperature_data, axis=1)

# Identify the month with the largest temperature variation
month_with_max_variation = np.argmax(monthly_temp_variation) + 1  # +1 to convert 0-index to 1-index for month
print("Month with the largest temperature variation:", month_with_max_variation)
```

```python
#Visualize the temperature data using line plots or heatmaps

import matplotlib.pyplot as plt

# Line plot for monthly average temperatures
plt.figure(figsize=(10, 6))
plt.plot(monthly_avg_temp, marker='o')
plt.title('Average Monthly Temperatures')
plt.xlabel('Month')
plt.ylabel('Average Temperature (Celsius)')
plt.xticks(ticks=np.arange(12), labels=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()

# Heatmap for all temperature data
plt.figure(figsize=(10, 6))
plt.imshow(temperature_data, aspect='auto', cmap='hot', interpolation='nearest')
plt.colorbar(label='Temperature (Celsius)')
plt.title('Monthly Temperature Data for All Stations')
plt.xlabel('Station')
plt.ylabel('Month')
plt.xticks(ticks=np.arange(5), labels=station_names)
plt.yticks(ticks=np.arange(12), labels=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
# Calculate daily returns

daily_returns = (stock_prices[1:] - stock_prices[:-1]) / stock_prices[:-1]
print("Daily returns:", daily_returns)
```

```python
# Calculate mean, standard deviation, and variance of daily returns

mean_return = np.mean(daily_returns)
std_dev_return = np.std(daily_returns)
variance_return = np.var(daily_returns)

print("Mean daily return:", mean_return)
print("Standard deviation of daily returns:", std_dev_return)
print("Variance of daily returns:", variance_return)
```

```python
# Identify the days with the highest and lowest returns

highest_return_day = np.argmax(daily_returns) + 1  # +1 to adjust for the shift caused by daily_returns calculation
lowest_return_day = np.argmin(daily_returns) + 1

print("Day with the highest return:", highest_return_day)
print("Day with the lowest return:", lowest_return_day)
```

```python
# Calculate cumulative returns

cumulative_return = np.prod(1 + daily_returns) - 1
print("Cumulative return over the year:", cumulative_return)
```

### Questin 03

You are working on a project that involves image processing. In this assignment, you will utilize NumPy to implement various operations on an RGB image of your own choice. Follow the steps below:

- Select an Image:
- Choose an RGB image of your own choice. You can use any image. Ensure that the image is in RGB format.
- Convert RGB Image to Grayscale:
- Convert the grayscale image to a binary image using a threshold value.
- Apply Gaussian blur to the binary image to reduce noise.
- Detect edges in the blurred image using the Sobel operator.
- Compute the histogram of pixel intensities in the original grayscale image.


<br> Visualize the results of each operation to observe the effects on the image.
Submission Guidelines:

<br>Submit the selected RGB image in the submission as well.

```python
import matplotlib.pyplot as plt
from PIL import Image
from skimage import color, filters
from scipy.ndimage import gaussian_filter, sobel

# Load the RGB image
image = Image.open("three_d_array.png")  # Replace "path_to_your_image" with the actual path to your RGB image file

# Convert the image to a NumPy array
three_d_array = np.array(image)

# Ensure the image is in RGB format (some images may have an alpha channel)
if three_d_array.shape[2] > 3:
    # Drop the alpha channel if present
    three_d_array = three_d_array[:, :, :3]

# Display the shape of the array to verify it's an RGB image
print("Shape of the RGB image array:", three_d_array.shape)
```

```python
# 2. Convert RGB Image to Grayscale

gray_image = color.rgb2gray(three_d_array)
```

```python
# Convert Grayscale Image to a Binary Image

threshold = 0.5  # Threshold value can be adjusted
binary_image = gray_image > threshold
```

```python
# Apply Gaussian Blur

blurred_image = gaussian_filter(binary_image.astype(float), sigma=1)
```

```python
# Detect Edges using Sobel Operator

edges = sobel(blurred_image)
```

```python
# Compute Histogram of Pixel Intensities in the Original Grayscale Image

hist, bins = np.histogram(gray_image, bins=256, range=(0, 1))

# Plotting the results
fig, ax = plt.subplots(2, 3, figsize=(15, 10))

ax[0, 0].imshow(three_d_array)
ax[0, 0].set_title('Original RGB Image')
ax[0, 0].axis('off')

ax[0, 1].imshow(gray_image, cmap='gray')
ax[0, 1].set_title('Grayscale Image')
ax[0, 1].axis('off')

ax[0, 2].imshow(binary_image, cmap='gray')
ax[0, 2].set_title('Binary Image')
ax[0, 2].axis('off')

ax[1, 0].imshow(blurred_image, cmap='gray')
ax[1, 0].set_title('Gaussian Blurred Image')
ax[1, 0].axis('off')

ax[1, 1].imshow(edges, cmap='gray')
ax[1, 1].set_title('Edges Detected (Sobel)')
ax[1, 1].axis('off')

ax[1, 2].plot(bins[:-1], hist)
ax[1, 2].set_title('Histogram of Grayscale Image')
ax[1, 2].set_xlabel('Pixel Intensity')
ax[1, 2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
