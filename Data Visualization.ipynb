### Data Loading and Inspection ###

```python
# Import Seaborn Library
import seaborn as sns
```

```python
# Load the Iris dataset from Seaborn
df = sns.load_dataset('iris')

```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Display the last few rows of the dataset
df.tail()
```

```python
# Print the summary statistics
df.describe()
```

```python
# check for any missing/null values in the dataset.
df.isnull().sum()
```

```python
# Display the datatype of each column.
df.dtypes
```

### Data Manipulation ###

```python
# Add a new column named 'sepal_area' that is the product of sepal length and sepal width.
df['sepal_area'] = df["sepal_length"] * df["sepal_width"]
df
```

```python
# Create a new DataFrame that contains only the data for the 'setosa' species.
setosa_df = df[df['species'] == 'setosa']
setosa_df.head()
```

```python
# Sort the original DataFrame by 'petal_length' in descending order
sorted_df = df.sort_values(by='petal_length', ascending=False)
print(sorted_df.head())
```

```python
# Filter the data to include only those records with a petal width greater than 1.5 cm.
filtered_df = df[df['petal_width'] > 1.5]
filtered_df.head()
```

```python
# Calculate the mean sepal width for each of the species present in the dataset.
mean_sepal_width = df.groupby('species')['sepal_width'].mean()
mean_sepal_width
```

### Data Visualization ###

```python
# Plot a histogram of the 'sepal_length' for the entire dataset. Customize the plot by adding a title and labels for x and y axes.
import matplotlib.pyplot as plt

# Plot histogram of 'sepal_length'
plt.figure(figsize=(8, 6))
plt.hist(df['sepal_length'], bins=15, color='skyblue', edgecolor='black')

# Add title and labels
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Create a scatter plot to visualize the relationship between 'sepal_length' and 'sepal_width'. Color the points by the species of the iris.
import seaborn as sns
import matplotlib.pyplot as plt

# Create scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='Set2')

# Add title and labels
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Show the plot
plt.grid(True)
plt.legend(title='Species')
plt.show()
```

```python
# Use a boxplot to show the distribution of 'petal_length' for each species. Customize this plot by adding a title and axis labels.
import seaborn as sns
import matplotlib.pyplot as plt

# Create boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='petal_length')

# Add title and labels
plt.title('Distribution of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Create a pairplot for all numerical columns, colored by species.
import seaborn as sns

# Create pairplot
sns.pairplot(df, hue='species', palette='Set2')

# Show the plot
plt.title('Pairplot of Iris Dataset')
plt.show()
```
